## 1: The Geometric Series

A *geometric progression* is a sequence generated by multiplying each term by some fixed number to get the next term
$$ a,\quad ar,\quad ar^2,\quad ar^3,\dots, . $$
The sum of $n$ terms of a geometric progression is
$$ S_n = \frac{a(1-r^2)}{1-r} $$

An *infinite series* is an expression in which an infinite number of terms are summed together. A *geometric series* is an infinite series whose terms form a geometric progression. A geometric series can be written in the form
$$ a + ar + ar^2 + \dots + ar^{n-1} + \dots . $$

The sum of a geometric series, defined as the limit of the sum of $n$ terms as $n\rightarrow\infty$, is
$$ S = \lim_{n\rightarrow\infty}S_n . $$

The sum of a geometric series exists if and only if $|r|<1$, and in this case is
$$ S = \frac{a}{1-r} . $$

### 1: Problems

**1.1**  Suppose a bouncing ball is originally dropped from a height of 1 yd and on each bounce rises 2/3 of the height of the previous bounce. Find the height of the tenth rebound, and the distance travelled by the ball after it touches the ground the tenth time. Compare this distance to the total distance travelled.

The heights of the successive bounces in yards can be represented by the geometric progression
$$ \frac{2}{3}, \frac{4}{9}, \frac{8}{27}, \dots, $$
which can be represented by the expression
$$ a, ar, ar^2, \dots, ar^{n-1}, \dots,$$
where $a = 2/3$ and $r = 2/3$.

Then the height of bounce $n$ can be expressed
$$ h_n = (2/3)\cdot(2/3)^{n-1} . $$
Then the height after ten bounces can be calculated

```python
a = 2/3 # initial term in geometric progression
r = 2/3 # common multiple in geometric progression
n = 10 # number of bounces

h_crit = a * r**(n-1) # height after n bounces

print('Height of the 10th rebound = ', h_crit, 'yd')
```

The distance $D(n)$ traveled by the ball after the tenth bounce can be determined from the following expression
$$ 
D(n) = 2 \cdot h_{10} + 2 \cdot h_{11} + \dots + 2 \cdot h_{n+9} + \dots \\
D(n) = 2(\cdot h_{10} + \cdot h_{11} + \dots \cdot h_{n+9} + \dots).
$$
The term in the parenthesis is a geometric series. Its sum $S$ therefore is 
$$ S = \frac{a}{1-r} , $$
where in this case $a = h_{10}$ and $r = 2/3$.

Then the distance travelled after the tenth bounce can be calculated

```python
a = h_crit
S = a / (1 - r)
D = 2 * S

print('Distance travelled after the 10th rebound = ', D, 'yd')
```
